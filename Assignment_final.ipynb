```python
# loading the companies data seperated by tab using latin1 encoding
# After trying different encoding schemes, group found that it works best with latin1
companies = pd.read_csv("companies.txt",delimiter="\t", encoding='latin1')
companies.head()
```

```python
# loading the rounds2 data seperated by default delimiter i.e. comma using latin1 encoding
# After trying different encoding schemes, group found that it works best with latin1
rounds2 = pd.read_csv("rounds2.csv", encoding="latin1")
rounds2.head()
```

```python
# Manual Inspection for companies dataframe
print("Columns : {}".format(rounds2.columns))
print("\nShape : {}".format(rounds2.shape))
print("\nVariable types : \n")
rounds2.info()
print("\nValues : {}".format(rounds2.values))
rounds2.describe()
```


### Merging the companies data with round2 into `master_frame`.


```python
#inner join is based on analysis requirement that is depends on combined data for both companies and rounds2
#from companies data related to country_code and category_list is essential
#from rounds2 to the raised_amount_usd and funding_round_type are improtant
# a new copy for the new merged dataframe is created
master_frame = pd.DataFrame(pd.merge(rounds2,companies, how='inner', 
                                     left_on = 'company_permalink', right_on = 'permalink')).copy()
master_frame.head()
```

```python
# Manual Inspection for merged master_frame dataframe
print("Columns : {}".format(master_frame.columns))
print("\nShape : {}".format(master_frame.shape))
print("\nVariable types : \n")
master_frame.info()
print("\nValues : {}".format(master_frame.values))
master_frame.describe()
```

```python
# As you can observe that in above data set there are few companies in rounds2 for which 
# there is no data set in companies data frame.
# the data is in same encoding latin1 for both companies and rounds2
# and contains non-english or special characters in permalink which couldn't be compared efficiently.
# On manual check it was also found that the count for such data is quiet less.
set(rounds2['company_permalink']).difference(set(companies['permalink']))
```

```python
# similar inspection for companies data not present in rounds2
set(companies['permalink']).difference(set(rounds2['company_permalink']))
```


### Start data cleaning process for `master_frame`


```python
# check the column-wise null count
master_frame.isnull().sum()
```

```python
# check he percentage of null data in each column of master_frame
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
# We are analyzing the investments done by others
# in absence of amount funded it is not relevant to take that data
#so this needs to be removed
# deleting rows with raised_amount_usd is Null
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
# for the investment to be performed the country code sould be known
# as one of the major analysis is based on the country and that too English speaking country
# therefore, the data with null values for country codes makes no sence
# and needs to be deleted from the master_frame
# deleting rows with  country_code is not present
master_frame = master_frame[~pd.isnull(master_frame['country_code'])]
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
#For the sector analysis also the data for category_list is required.
# If case this data is missing, investment decision cannot happen
# So deleting the rows from the masterframe where category_list is not present
master_frame = master_frame[~pd.isnull(master_frame['category_list'])]
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
# dropping the non-required columns after reviewing what is required across all checkpoints
# company_permalink is dropped as it was duplicate
#funding_round_code,homepage_url,state_code,region,city,founded_at are dropped as not required for analysis
dropping_columns = ['company_permalink','funding_round_code','homepage_url','state_code','region','city','founded_at']
master_frame = master_frame.drop(dropping_columns,axis=1)
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
# final master_frame stats for null data after data cleaning
# all percentages as 0 suggest that the data is finally cleaned and can be used for apprpriate analysis
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
# check the master_frame as how many number of rows and percentage of rows that have been retained.
print("number of rows present now : {}".format(len(master_frame.index)))      
percent_rows_retained = round(100*(len(master_frame.index)/114942),2)
print("percent rows retained : {}".format(percent_rows_retained))
```

# Checkpoint 2: Funding Type Analysis

```python
# As part of Funding type analysis we are required to compare the amounts in million usd.
# however the amounts mentioned in raised_amount_usd column is in usd and mentioned in exp. form
# this is dufficult to compare
# therefore, a new column raised_amount_million_usd is added to master_frame that contains million dollars
# also the decimal places are taken till 3 as it is based on international counting standard
master_frame['raised_amount_million_usd'] = master_frame['raised_amount_usd'].apply(lambda x: round(x/1000000,3))
master_frame.head()
```

```python
# as per requirement we are grouping the master_frame based on funding_round_type
# on the group, we are applying the mean aggrgated function
# to get the mean values in raised_amount_million_usd
master_frame_groupby = master_frame.groupby('funding_round_type')
master_frame_groupby.mean()
```

### Conclusion : On checking above result, it is observed that between 5 to 15 million USD there is only one funding round type which is for `venture` funding type

```python
# creating the dataframe copy for the most applicable funding_round_type "venture"
# which can be used further in later analsis
venture_frame = pd.DataFrame(master_frame.loc[master_frame.funding_round_type=='venture']).copy()
venture_frame.head()
```

# Checkpoint 4: Sector Analysis 1


```python
#After reviewing the data we can see that the funding is happening across a lot of sectors
#Lets divide them into 8 main sectors to make our task easy
#For that, first we need to know which one is the primary sector
# Extracting primary sector among category list into a new column Primary_Sector
master_frame['Primary_Sector'] = master_frame['category_list'].str.split("|", expand = True)[0]

#Check to make sure that values are populated correctly
master_frame[['category_list','Primary_Sector']].head()
```

```python
#Now, we know the primary sector. Lets find out which is the main sector.
# For that, lets first import the data of mapping file.
mapping = pd.read_csv("mapping.csv", encoding='latin1')
mapping.head()
```

```python
# Manual Inspection for merged mapping dataframe
print("Columns : {}".format(mapping.columns))
print("\nShape : {}".format(mapping.shape))
print("\nVariable types : \n")
mapping.info()
print("\nValues : {}".format(mapping.values))
mapping.describe()
```

```python
#On manual analysis, it was noticed that at some places instead of word na 0 is written like Analytics 
# is written as A0lytics. Lets change it first. Otherwise we wont be able to recode all the sectors
mapping['category_list'].replace(to_replace = '0',value = 'na', inplace = True,  regex=True)
#Execution of above statement also changes Enterprise 2.0 to Enterprise 2.na which is not needed. 
#So, lets change it back to its original value.
mapping['category_list'].replace(to_replace = 'Enterprise 2.na',value = 'Enterprise 2.0', inplace = True,  regex=True)
#Check if the recoding is done properly. You can see that now A0lytics is indeed Analytics
mapping['category_list'].head(50)
```

```python
# It is observed that the mapping dataframe is not straight forward to extract the main sectors.
# In order to extract the main sectors we need to prepare a seperate dataframe that provides the
# apropriate mapping for category_list and main_sectors.
# This new dataframe called melted_dataframe is prepared by using melt function.
# the result dataframe contains category_list, main_sector 
# and sec_value(0 or 1 as mentioned in original mapping dataframe).
cols =  list(mapping.columns)
cols.remove('category_list')
#To make it simple lets first change the data from wide format to long format using melt function
melted_mapping = mapping.melt(id_vars = 'category_list', 
                        value_vars = cols, var_name='main_sector', value_name = 'sec_value')
melted_mapping.head(20)
```

```python
# among the melted_mapping we require only the mappings where the sec_value is 1
# a value of sec_value=0 means that the category_list doesn't belong to the main_category
# after filtering the required values we shall be left with appropriate dataframe
# containing strainght forward and useable mappings for category_list and main_sector
melted_mapping = melted_mapping[melted_mapping['sec_value']==1]
melted_mapping.head(50)
```

```python
#Perform the merging of master_frame and mapping database to populate the main_sector in master_frame
# inner join is taken so that blanks values which cannot be mapped to any of  
# the main sectors in mapping table do not appear in the result
master_frame = pd.merge(master_frame, melted_mapping[['main_sector']], 
                        how='inner', 
                        left_on = master_frame.Primary_Sector.str.lower(), 
                        right_on = melted_mapping.category_list.str.lower())

#Check the data after merging to make sure it is populated correctly
master_frame.head()
```
